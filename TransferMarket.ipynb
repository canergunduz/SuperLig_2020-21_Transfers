```python
from selenium import webdriver
import bs4 as bs
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
player_list, from_team_list, from_league_list, to_team_list, to_league_list= list(), list(),list(), list(),list()
```

```python
#driver = webdriver.Chrome("C:\\chromedriver.exe")
driver.get("https://www.transfermarkt.com/transfers/saisontransfers/statistik?land_id=0&ausrichtung=&spielerposition_id=&altersklasse=&leihe=&transferfenster=&saison-id=2020&plus=1")
```

```python
def sayfa_yukle(driver):
    global player_list
    global from_team_list
    global from_league_list
    global to_team_list
    global to_league_list
    driver.set_page_load_timeout(30)
    #driver.maximize_window()
    driver.implicitly_wait(20)
    aa = driver.find_elements_by_xpath("//*[(@id = '"'yw0'"')]//td")
    for a in range(len(aa)):
        tds=bs.BeautifulSoup(aa[a].get_attribute("innerHTML"), "html.parser").find_all()
        if a % 17==3:
            try:
                player_list.append(tds[0].contents)
            except IndexError:
                player_list.append("")
        elif a % 17==10:
            try:
                from_team_list.append(tds[0].contents)
            except IndexError:
                from_team_list.append("")
        elif a % 17==11:
            try:
                from_league_list.append(tds[1].contents)
            except IndexError:
                from_league_list.append("")
        elif a % 17==14:
            try:
                to_team_list.append(tds[0].contents)
            except IndexError:
                to_team_list.append("")
        elif a % 17==15:
            try:
                to_league_list.append(tds[1].contents)
            except IndexError:
                to_league_list.append("")
    
    driver.find_element_by_xpath('//*[@title="Go to next page"]').click()
    return driver
```

```python
import time 
        
while driver.find_element_by_xpath('//*[@title="Go to next page"]') != "":
    time.sleep(7)
    sayfa_yukle(driver)
```

```python
df1 = pd.DataFrame(player_list,columns=['Player'])
df2 = pd.DataFrame(from_team_list,columns=['From_Team'])
df3 = pd.DataFrame(from_league_list,columns=['From_League'])
df4 = pd.DataFrame(to_team_list,columns=['To_Team'])
df5 = pd.DataFrame(to_league_list,columns=['To_League'])
```

```python
result = pd.concat([df1, df2, df3, df4, df5], axis=1)
```
